## โจทย์เรื่องการตั้งค่า Variable
จงเลือกตอบ <br>
YES ถ้าข้อความต่อไปนี้ถูกต้องตามหลักการตั้งตัวแปร <br>
NO ถ้าข้อความต่อไปนี้ไม่ถูกต้องตามหลักการตั้งตัวแปร และตอบว่าเพราะอะไร

1. > name = 'ชื่อของคุณ'

ตอบ: YES

2. > _name = 'ชื่อของคุณ'

ตอบ: YES

3. > name1 = 'ชื่อของคุณ'

ตอบ: YES

4. > name! = 'ชื่อของคุณ'

ตอบ: NO เพราะมีอักขระพิเศษ!

5. > !name = 'ชื่อของคุณ'

ตอบ: NO เพราะมีอักขระพิเศษ!

6. > number = 20

ตอบ: YES

7. > _number = 20

ตอบ: YES

8. > @number = 20

ตอบ: NO เพราะมีอักขระพิเศษ!

9. > number@ = 20

ตอบ: NO เพราะมีอักขระพิเศษ!

10. > a5 = 20

ตอบ: YES

11. > 5a = 20

ตอบ: NO ชื่อตัวแปรไม่สามารถเริ่มต้นด้วยตัวเลขได้

12. > favorite_food = 'Hamburger'

ตอบ:  YES

13. > FavoriteFood = 'Hamburger'

ตอบ: YES

14. > favorite food = 'Hamburger'

ตอบ: NO ตัวแปรไม่สามารถเว้นวรรคได้

15. > if = 7

ตอบ: NO มีคำสงวน if

16. >else = 5

ตอบ: NO มีคำสงวน else

17. >for = 6

ตอบ:  NO มีคำสงวน for


18. > a=5, b=6, c=7

ตอบ: NO ผิดรูปแบบต้องใช้เครื่องหมาย ; ดังนั้น a=5; b=6; c=7

19. > a, b, c = 5, 6, 7

ตอบ: YES

20. > a=5; b=6; c=7

ตอบ: YES
