### การใช้ print() รูปแบบ ผสม ทั้ง String และ Numeric

กำหนดให้ a = 2
19. > a = 2

```python
a = 2
print("a = %d" %a)
```

กำหนดให้  a = 2, b = 5
20. > a = 2 <br>
> b = 5 <br>
> 2 x 5 = 6






```python
a = 2
b = 5
print("""
a = %d\n
b = %d\n
%d x %d = %d
""" %(a, b, a, b, (a*b)))
```

กำหนดให้ a = 2, b = 3, c = 5
21. > a = 2<br>
> b = 3<br>
> c = 5<br>
> 2*(3+5) = 2 * 3 + 2 * 5

```python
a = 2
b = 3
c = 5
print("""
a = %d\n
b = %d\n
c = %d\n
%d*(%d+%d) = %d*%d + %d*%d
""" %(a, b, c, a, b, c, a, b, a, c))
```

กำหนดให้ค่าในตัวแปร day month year
22. > ฉันเกิดวันที่ 1 มกราคม 2077
