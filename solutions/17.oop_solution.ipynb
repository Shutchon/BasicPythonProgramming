method ดังนี้

```
attribute
ㆍr เป็นค่าของส่วนจริง
ㆍi เป็นค่าของส่วนจินตภาพ

method
ㆍget_value() ทำหน้าที่แสดงค่า canesian form เช่น 2+3i, 5-4i เป็นต้น
โดยเมื่อเรียกใช้ get value() โปรแกรมจะพิมพ์ <r> + <i>i
```

###ตัวอย่าง
input
```
Z = ComplexNumber(2, 4)
Z.get_value()
```

output
```
2+4i
```

````python
# for the class
class ComplexNumber:
  def __init__(self, real, imaginary) -> None:
    self.r = real
    self.i = imaginary

  def get_value(self):
    if self.i > 0:
      print(f"{self.r}+{self.i}i")
    if self.i < 0:
      print(f"{self.r}{self.i}i")
    if self.i == 0:
      print(f"{self.r}")
````

```
attribute

method
ㆍ get_area() ทำหน้าที่แสดงค่าพื้นที่ของวงกลม (area = pi*radius**2)
ㆍ get_perimeter() ทำหน้าที่แสดงค่าความยาวเส้นรอบวง (perimeter = 2*pi*radius)
```

###ตัวอย่าง
input
```

circle_B = Circle(16)
print('Area Circle A :', circle_A.get_area())
print('Perimeter Circle A :', circle_A.get_perimeter())
```

````python
# for the class
import math
class Circle:

  def __init__(self, r):
    self.r = r

  def get_area(self):
    return math.pi*self.r**2

  def get_perimeter(self):
    return 2*math.pi*self.r
````

```
ㆍbook_status เป็นสถานะการยืมของหนังสือ โดย book_status มีค่า
default เป็น 'available'

method
ㆍborrow_book()
  - ถ้า book_status เป็น 'available'
    * ให้พิมพ์ 'borrow successfully'
    * เปลี่ยน book status เป็น 'unavailable'
  - แต่ถ้า เรียกใช้ตอนที่ book status เป็น 'unavailable'
    * ให้พิมพ์ 'borrow unsuccessfully'
ㆍ return_book()
  - ถ้า book_status เป็น 'unavailable'
    * ให้พิมพ์ 'return successfully'
    * เปลี่ยน book_status เป็น 'available'
  - แต่ถ้า เรียกใช้ตอนที่ book status เป็น 'available'
    * ให้พิมพ์ 'return unsuccessfully'
```

###ตัวอย่าง
input
```
Book_A = Book ('AI Superpowers')
print('Book status :', Book_A.book_status)
Book_A.borrow_book ()
print('Book status :', Book_A.book_status)
Book_A.return_book ()
print('Book status :', Book_A.book_status)
```

output
```
Book status : available
borrow successfully
Book status : unavailable
return successfully
Book status : available
```

````python
# for the class
class Book:
  def __init__(self, book_name):
    self.name = book_name
    self.status = 'available'

  def borrow_book(self):
    if self.status == 'available':
      print("borrow successfully")
      self.status = 'unavailable'
    else:
      print("borrow unsuccessfully")

  def return_book(self):
    if self.status == 'unavailable':
      print("return successfully")
      self.status = 'available'
    else:
      print("return unsuccessfully")

````

````python
# for the running
Book_A = Book ('AI Superpowers')
print('Book status :', Book_A.status)
Book_A.borrow_book ()
Book_A.borrow_book ()
print('Book status :', Book_A.status)
Book_A.return_book ()
Book_A.return_book ()
print('Book status :', Book_A.status)
````

8. เขียนโปรแกรมสร้าง class ชื่อ Stats โดยมี attibute และ method ดังนี้

```
attribute
ㆍdata เป็น list ที่เก็บจำนวนใด ๆ

method
ㆍ mean() ทำหน้าที่คำนวณค่าเฉลี่ยของจำนวนใน data
ㆍ min() ทำหน้าที่คำนวณค่าต่ำสุดของจำนวนวนใน data
ㆍ max() ทำหน้าที่คำนวณค่าสูงสุดของจำนวนใน data
ㆍ summary() ทำหน้าที่แสดงค่า mean, min และ max โดยเมื่อเรียกใช้
โปรแกรมจะพิมพ์ Mean : <mean>, Min : <min> และ Max: <max> ออกมา
```

###ตัวอย่าง
input
```
A = [1,5,7,9,11]
stat_A = Stats(A)
stat_A.summary()
```

output
```
Mean : 6.6
Min : 1
Max : 11
```

````python
# for the class
import statistics
class Stats:
  def __init__(self, data_list):
    self.data = data_list

  def list_mean(self):
    # sum = 0
    # for i in self.data:
    #   sum = sum + i
    # return sum/(len(self.data))
    return statistics.mean(self.data)

  def list_max(self):
    # max = float('-inf')
    # for i in self.data:
    #   if i > max :
    #     max = i
    # return max
    return max(self.data)

  def list_min(self):
    # min = float('inf')
    # for i in self.data:
    #   if i < min :
    #     min = i
    # return min
    return min(self.data)

  def summary(self):
    print("Mean: ", self.list_mean())
    print("Max: ", self.list_max())
    print("Min: ", self.list_min())
````

````python
# for the running
A = [1,5,7,9,11]
stat_A = Stats(A)
stat_A.summary()
````

```
attribute
ㆍ amount เป็นจำนวนเงิน
ㆍ currency เป็นสกุลเงิน

method
ㆍ to_THB() ทำหน้าที่แปลงสกุลเงินไปเป็น THB
ㆍ to_USD() ทำหน้าที่แปลงสกุลเงินไปเป็น USD
(กำหนดให้ใช้อัตราแลกเปลี่ยน 1 USD = 30 THB)
```

###ตัวอย่าง
input
```
money=USDTHB(300000, 'USD')
print(money.amount, money.currency)
money.to_THB()
print(money.amount, money.currency)
money.to_USD()
print(money.amount, money.currency)
```

output
```
30000 USD
900000 THB
30000.0 USD
```

````python
# for the class
class USDTHB:
  def __init__(self, amount, currency):
    self.amount = amount
    self.currency = currency

  def to_THB(self):
    if self.currency == 'USD':
      self.currency = 'THB'
      self.amount = self.amount * 30
    else:
      self.amount = self.amount

  def to_USD(self):
    if self.currency == 'THB':
      self.currency = 'USD'
      self.amount = self.amount / 30
    else:
      self.amount = self.amount
````

````python
# for the running
money=USDTHB(300000, 'USD')
print(money.amount, money.currency)
money.to_THB()
print(money.amount, money.currency)
money.to_USD()
print(money.amount, money.currency)
````

10. เขียนโปรแกรมสร้าง class ชื่อ Cashier โดยมี attribute และ method ดังนี้

```
attribute
ㆍproducts เป็น list ที่เก็บชื่อสินค้า

method
ㆍ recommend() ทำหน้าที่แนะนำสินค้า โดยเมื่อเมื่อเรียกใช้ โปรแกรมจะ
พิมพ์ We have <product>. ออกมา
```

###ตัวอย่าง
input
```
cashier = Cashier (['apple', 'banana', 'orange'])
cashier.recommend()
```

output
```
We have apple, banana, orange.
```

````python
# for the class
class Cashier:
  def __init__(self, product_data):
    self.products = product_data

  def recommend(self):
    single_products = []
    for i in self.products:
      if i not in single_products:
        single_products.append(i)

    textp = ""
    for i in single_products:
      textp = textp + i + ", "

    textp = textp[:-2] # เพื่อตัด ", " ตัวสุดท้าย
    text = f"We have {textp}."

    return text
````

````python
# for the running
cashier = Cashier (['apple', 'banana', 'orange'])
cashier.recommend()
````

```
ㆍ attack(warrior_object) ทำหน้าที่แสดงค่าพลังชีวิตหลังจากถูกโจมตี
โดยที่
  - ถ้า power < defense แล้ว HP จะไม่ลด
  - ถ้า power > defense แล้ว HP จะมีค่าเท่ากับ HP-(power-defense)
โดยเมื่อเรียกใช้ โปรแกรมจะแสดงค่า HP ออกมา และถ้า HP = 0 ให้
```

###ตัวอย่าง

```
Warrior_B.attack (Warrior_B)
Warrior_B.attack (Warrior_A)

```

```
Warrior_A HP = 80
Warrior_B HP = 120
=== After Attack ===
Warrior_A HP = 70
Warrior_B HP = 100
```

````python
# for the class
class Warrior:
  def __init__(self, power, defend, hp):
    self.HP = hp
    self.power = power
    self.defend = defend

  def attack(self, EnemyWarrior):
    if self.power > EnemyWarrior.defend:
      EnemyWarrior.HP = EnemyWarrior.HP - (self.power - EnemyWarrior.defend)
      if EnemyWarrior.HP <= 0:
        print("Enemy died")


````

````python
# for the running
Warrior_A = Warrior(100,50,80)
Warrior_B = Warrior(60, 80, 120)
print('=== Before Attack ===')
print('Warrior_A HP =', Warrior_A.HP)
print('Warrior_B HP =', Warrior_B.HP)
Warrior_B.attack(Warrior_B)
Warrior_B.attack(Warrior_A)
print('=== After Attack ===')
print('Warrior_A HP =', Warrior_A.HP)
print('Warrior_B HP =', Warrior_B.HP)
Warrior_B.attack(Warrior_A)
Warrior_B.attack(Warrior_A)
Warrior_B.attack(Warrior_A)
Warrior_B.attack(Warrior_A)
Warrior_B.attack(Warrior_A)
Warrior_B.attack(Warrior_A)
Warrior_B.attack(Warrior_A)
Warrior_B.attack(Warrior_A)
Warrior_B.attack(Warrior_A)
Warrior_B.attack(Warrior_A)
````

```
  - เลื่อนตำแหน่งไปทางซ้ายตามแนวแกน x ไป d หน่วย
  - พิมพ์ Current position : (<x>, <y>) ออกมา
ㆍmove_right(d) รับ parameter 1 ตัว คือ d
  - เลื่อนตำแหน่งไปทางขวาตามแนวแกน x ไป d หน่วย
  - พิมพ์ Current position : (<x>, <y>) ออกมา
```

###ตัวอย่าง

```
A = Point(0, 0)
```


````python
# for the class
class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def move_up(self, d):
    self.y = self.y + d
    self.current_pos()

  def move_down(self, d):
    self.y = self.y - d
    self.current_pos()

  def move_left(self, d):
    self.x = self.x - d
    self.current_pos()

  def move_right(self, d):
    self.x = self.x + d
    self.current_pos()

  def current_pos(self):
    print(f"Current position : ({self.x},{self.y})")
````

13. เขียนโปรแกรมสร้าง class ชื่อ Businessman โดยมี attribute และ method

```
attribute

ㆍmoney เป็นพลังการเงิน
ㆍhappiness เป็นพลังความสุข

method
ㆍwork() ทำหน้าที่แสดงค่าพลังชีวิต การเงิน และความสุข หลังจากผ่าน
การทำงาน โดยจะเพิ่ม money 20 หน่วย และ happiness 10 หน่วย
แต่จะลด HP 10 หน่วย
ㆍsleep() ทำหน้าที่แสดงค่าพลังชีวิต การเงิน และความสุข หลังจากผ่าน
การนอนหลับ โดยจะเพิ่ม HP 20 หน่วย และ happiness 10 หน่ว
จะลด money 10 หน่วย
ㆍplay() ทำหน้าที่แสดงค่าพลังชีวิต การเงิน และความสุข หลังจากผ่าน
การเล่นเกม โดยจะลด HP และ money 10 หน่วย แต่จะเพิ่ม happi
ness 20 หน่วย
ㆍreport() ทำหน้าที่แสดงค่าพลังชีวิต การเงิน และความสุข
```

###ตัวอย่าง
input
```
John = Businessman(100,100,100)
John.work()
John.report()
John.play()
John.report()
John.sleep()
John.report()
```

output
```
HP = 100, Money = 100, Happiness = 100
HP = 90, Money = 120, Happiness = 110
HP = 80, Money = 110, Happiness = 130
HP = 100, Money = 100, Happiness = 140
```




````python
# for the class
class Businessman:
  def __init__(self, hp, money, happiness):
    self.HP = hp
    self.money = money
    self.happiness = happiness
    self.report()

  def work(self):
    self.HP -= 10
    self.money += 20
    self.happiness += 10

  def play(self):
    self.HP -= 10
    self.money -= 10
    self.happiness += 20

  def sleep(self):
    self.HP += 20
    self.money -= 10
    self.happiness += 10

  def report(self):
    print(f"HP = {self.HP}, Money = {self.money}, Happiness = {self.happiness}")
````

````python
# for the running
John = Businessman(100,100,100)
John.work()
John.report()
John.play()
John.report()
John.sleep()
John.report()
````
