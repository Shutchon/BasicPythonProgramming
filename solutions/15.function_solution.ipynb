```
สมมุติ str1 = '@chicken' จะ return 'chicken'
สมมุติ str1 = '1,2%3$' จะ return '123'
```

````python
# สำหรับเขียน Function
def remove_symbol(str1):
  result = ''
  for i in str1:
    if i not in '!@#\\$%,':
      result += i
  return result
````

1 ตัว ได้แก่ amount ซึ่งเป็นจำนวนเต็มบวกใด ๆ ให้ return dictionary ที่มี
key เป็น '10', '5', '2', '1'และ value เป็นจำนวนเหรียญชนิดนั้น ๆ โดยจะต้อง
แลกเหรียญที่มีค่ามากให้เสร็จก่อน

```
สมมติให้ amount = 40 จะ return {'10':4,'5':0,'2':0,'1':0}
สมมติให้ amount = 48 จะ return {'10':4,'5':1,'2':1,'1':1}
```

````python
# สำหรับเขียน Function
def give_change(money): #amount
  if money <= 0:
    return 0

  result10=0
  result5=0
  result2=0
  result1=0
  result10 = money//10
  money = money - (result10*10) # money = money%10
  result5 = money//5
  money = money - (result5*5)
  result2 = money//2
  money = money - (result2*2)
  result1 = money//1

  return {'10':result10,'5':result5,'2':result2,'1':result1}

````

````python
# สำหรับ Run
print(give_change(48))
````

25. เขียนฟังก์ชันเพื่อหาระยะห่างระหว่างจุด 2 จุดในระนาบ n มิติ และ return
ผลลัพธ์ออกมา (hint: euclidean distance)
กำหนดให้ฟังก์ชันนี้รับ parameter 2 ตัว คือ P1 และ P2 ที่เป็น tuple ความยาว n

````python
# สำหรับเขียน Function
import math
def euclidean_distance(P1, P2):
  #P1 = (X1,Y1) , P2 = (X2,Y2)
  result = math.sqrt((P2[0]-P1[0])**2+(P2[1]-P1[1])**2)
  return round(result, 2)
````

````python
# สำหรับ Run
P1 = (0,0)
P2 = (5,10)
print(euclidean_distance(P1,P2))
````

27. เขียนฟังก์ชันสำหรับ shift สมาชิกใน list ไปทางซ้าย k step โดยรับ parame-
ter 2 ตัว ได้แก่ listA ซึ่งเป็น list ที่เก็บข้อมูลใด ๆ และ k ซึ่งเป็นจำนวนครั้งที่ต้องการให้ shift ไปทางซ้าย ให้ return listA ที่ผ่านการ shift ออกมา

```
สมมติให้ listA = [1, 2, 3, 4, 5] และ k = 1 จะ return [2, 3, 4, 5, 1]
สมมติให้ listA = [1, 2,3, 4, 5] และ k = 2 จะ return [3, 4, 5, 1, 2]
```

````python
# สำหรับเขียน Function
def shitf_left(listA, k):
  result = [0]*len(listA)
  for i in range(len(listA)):
    result[(i-k)%len(listA)] = listA[i]
  return result

````

````python
# สำหรับ Run
listA = [1, 2, 3, 4, 5]
print(shitf_left(listA, 1))
print(shitf_left(listA, 2))
````

31. เขียนฟังก์ชันเพื่อหาค่าต่ำสุด และค่าสูงสุด ให้ return ผลลัพธ์ทั้ง 2 ออกมา กำหนดให้ฟังก์ชันนี้รับ parameter 1 ตัว คือ listA ที่มีสมาชิกเป็นจำนวนใด ๆ

*hint:* return เป็น list หรือ tuple

````python
# สำหรับเขียน Function
def min_max(listA):
  min_result=float('inf')
  max_result=float('-inf')
  for i in listA:
    if i < min_result:
      min_result = i
    if i > max_result:
      max_result = i
  return (min_result, max_result)

````

````python
# สำหรับ Run
listA =  [1,8,6,88,4,5,99,22,45,-6]
print(min_max(listA))
````

```
ㆍถ้าเป็นจำนวนเฉพาะ ให้ return True
ㆍถ้าไม่เป็นจำนวนเฉพาะ ให้ return False
```


````python
# สำหรับเขียน Function
def is_prime_number(num):
  num = abs(num)
  for i in range(2,num-1):
    if num % i == 0:
      print(i)
      return False
  return True
````

````python
# สำหรับ Run
print(is_prime_number(117))
````

35. เขียนฟังก์ชันสำหรับวาครูปต่อไปนี้โดยรับ parameter 1 ตัว คือ n

```
n=5
*---*
-*-*-
--*--
-*-*-
*---*
```

```
n=6
*----*
-*--*-
--**--
--**--
-*--*-
*----*
```


````python
# สำหรับเขียน Function
def draw(n):
  for i in range(n):
    line = ""
    for j in range(n):
      if (j == 0+i) or (j == (n-1)-i):
        line+='*'
      else:
        line+='-'
    print(line)

````

````python
# สำหรับ Run
draw(6)
````
